```python
# ÿßŸÜÿ™ÿÆÿßÿ® ÿ≥ÿ™ŸàŸÜ‚ÄåŸáÿß€å ÿßÿπÿ¥ÿßÿ±€å
float_df = df.select_dtypes(include='float')

# ÿßŸÜÿ™ÿÆÿßÿ® ÿ≥ÿ™ŸàŸÜ‚ÄåŸáÿß€å ÿπÿØÿØ ÿµÿ≠€åÿ≠
int_df = df.select_dtypes(include='int')

# ÿßÿØÿ∫ÿßŸÖ ÿØŸà ÿØ€åÿ™ÿßŸÅÿ±€åŸÖ ÿ®ÿß ŸáŸÖ
num_df = pd.concat([int_df, float_df], axis=1)
```

```python
# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ŸÖŸÇÿßÿØ€åÿ± ÿ¢ŸÖÿßÿ±€å ÿØÿßÿØŸá‚ÄåŸáÿß
from tabulate import tabulate

summary_df = pd.DataFrame(columns=["Column", "Mean", "Median", "Standard Deviation", "Variance", "Trimmed Mean", "Skewness", "Kurtosis", "Mode", "Correlation"])

for col in num_df.columns:
    mean = num_df[col].mean()
    median = num_df[col].median()
    std = num_df[col].std()
    var = num_df[col].var()
    trim_min = num_df[col].quantile(0.1)
    trim_max = num_df[col].quantile(0.9)
    trimmed_mean = num_df[col][(num_df[col] >= trim_min) & (num_df[col] <= trim_max)].mean()
    skew = num_df[col].skew()
    kurt = num_df[col].kurtosis()
    mode = num_df[col].mode().values[0]
    corr = num_df.corr()[col].to_list()[:-1] # ÿ±ÿßÿ®ÿ∑Ÿá ÿ®€åŸÜ ÿØŸà ÿ≥ÿ™ŸàŸÜ ÿ®ÿ±ÿß€å ÿ≥ÿ™ŸàŸÜ ŸÅÿπŸÑ€å
    summary_df = pd.concat([summary_df, pd.DataFrame({
        "Column": [col],
        "Mean": [mean],
        "Median": [median],
        "Standard Deviation": [std],
        "Variance": [var],
        "Trimmed Mean": [trimmed_mean],
        "Skewness": [skew],
        "Kurtosis": [kurt],
        "Mode": [mode],
        "Correlation": [corr]
    })], ignore_index=True)

# ÿ™ÿ®ÿØ€åŸÑ ŸÖŸÇÿßÿØ€åÿ± ÿ®Ÿá ÿ¨ÿØŸàŸÑ ÿ≤€åÿ®ÿß
table = tabulate(summary_df, headers="keys", tablefmt="fancy_grid")

# ⁄ÜÿßŸæ ÿ¨ÿØŸàŸÑ
print(table)
```

```python
def draw_barplot(_style, _x,_y,_dataset,_suptitle,_xLabels,_yLabels):        
            sns.set(style=_style)         
            g = sns.catplot(x=_x, y=_y, data=_dataset,aspect=2,
                            height=5, kind="bar", palette="muted")
            plt.subplots_adjust(top=0.9)
            g.fig.suptitle(_suptitle)
            g.set_xlabels(_xLabels)
            g.set_ylabels(_yLabels)    
```

```python
def plot_CountPlot(X,Dataframe,Title):
    fig = plt.figure(figsize=(15,5))
    ax = sns.countplot(x=X, data=Dataframe ,palette='pastel' ,edgecolor=sns.color_palette("dark", 3))
    ax.set_title(Title)
    ax.legend(loc='upper right')

    for t in ax.patches:
        if (np.isnan(float(t.get_height()))):
            ax.annotate(0, (t.get_x(), 0))
        else:
            ax.annotate(str(format(int(t.get_height()), ',d')), (t.get_x(), t.get_height()*1.01))
    plt.show();   
```

```python
dt=df.groupby(by='origin').count().reset_index().sort_values(by='year', ascending=False)
dt['flightsCount']=dt.year

#Call Function to draw Bar Graph
draw_barplot(_style='whitegrid',_x='origin',_y='flightsCount',_xLabels='Airport',
             _yLabels='Number of Flights', _dataset=dt,
             _suptitle="Number of Flights from different Airports"
            )
```

```python
plot_CountPlot('origin',df,'Number of Flights from different Airports')
```
